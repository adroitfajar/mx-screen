```python
from pathlib import Path
import numpy as np
from ase.io import read, write
from ase.build import make_supercell

# Paths
IN_DIR  = Path("cartoon")          # folder with *_cartoon.xyz
OUT_DIR = Path("periodic")         # output folder
OUT_DIR.mkdir(exist_ok=True)

# Parameters (visual only)
VACUUM_Z = 25.0     # Å
P = np.diag([2, 2, 1])   # 2×2 in-plane replication

# Loop over all cartoon.xyz
xyz_files = sorted(IN_DIR.glob("*_cartoon.xyz"))

print(f"Found {len(xyz_files)} cartoon structures")

for xyz in xyz_files:
    atoms = read(xyz)

    # Determine fake lattice from atomic extents
    pos = atoms.get_positions()
    lx = pos[:, 0].max() - pos[:, 0].min() + 3.0
    ly = pos[:, 1].max() - pos[:, 1].min() + 3.0
    lz = VACUUM_Z

    atoms.set_cell([lx, ly, lz])
    atoms.center()

    # 2×2 replication
    atoms_2x2 = make_supercell(atoms, P)

    # Output names
    stem = xyz.stem.replace("_cartoon", "")
    cif_out = OUT_DIR / f"{stem}_2x2_visual.cif"
    xyz_out = OUT_DIR / f"{stem}_2x2_visual.xyz"

    # Write files
    write(cif_out, atoms_2x2)
    write(xyz_out, atoms_2x2)

    print(f"✔ {stem}: saved 2×2 slab")

print("\nDone. Open *_2x2_visual.cif in VESTA.")
```
